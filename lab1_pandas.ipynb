```python
funcs = [np.nanstd, np.nanmean, np.var ]
```

```python
computed_metrics = df.groupby('company').agg({'extraction_fact' :funcs, 'extraction_plan' : funcs,
                           'ash_percent_fact' : funcs, 'ash_percent_plan' : funcs})
```

<pre style="font-size : 25pt; text-align : center"><b>ACTUAL PLOTS</b></pre>

```python
def make_trend_line(x,y,degree):
    return np.poly1d(np.polyfit(x,y,degree))
```

```python
def make_barplot(first_field='extraction_fact', second_field='nanmean', degree=4):
    plt.figure(figsize=(15,12))
    plt.title(first_field, fontdict={"size": 20, "family" : "serif"})
    plt.xlabel("companies",fontdict={'size' :20, 'family':"serif"})
    plt.ylabel('Metric : {}'.format(second_field), fontdict={'size' :20, 'family':"serif"})
    x = list(computed_metrics.index)
    y= computed_metrics[first_field][second_field]
    plt.bar(range(len(x)), y);
    plt.xticks(range(len(x)),x, rotation=35);
    plt.yticks(np.arange(0, max(y), int(min(y))/3));
    trend = make_trend_line(range(len(x)), y.values, degree=degree)
    plt.plot(x,trend(range(len(x))), '--', color='red')
    plt.legend(['Line of trend (Linear regression with polynomial features)'])
```

```python
make_barplot()
```

```python
make_barplot("extraction_plan")
```

```python
make_barplot("ash_percent_fact")
```
